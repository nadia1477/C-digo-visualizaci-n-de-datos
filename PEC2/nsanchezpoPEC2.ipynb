```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
from pandas import DataFrame
pd.set_option('display.max_columns', None)
```

Cargar dataset


```python
df = pd.read_csv('pax_data_1789_agreements_08-11-19.csv')
print("número de filas:", df.shape[0], "número de columnas:", df.shape[1])
```

Valores perdidos


```python
df.isnull().sum() 
df.isna().sum()
```

```python
df.dropna(axis=1, how='all', inplace=True)
df.isnull().sum()/len(df)*100
#eliminar variables con mas del 5% de valores NaN
delete_columns = df.columns[df.isnull().sum()/len(df)*100 > 5]
delete_columns
```

```python
df2= DataFrame(df.drop(delete_columns, axis=1))
print("número de filas:", df2.shape[0], "número de columnas:", df2.shape[1])
```

```python
df3= DataFrame(df2.dropna())
print("número de filas:", df3.shape[0], "número de columnas:", df3.shape[1])
```

```python
# Crear matriz de correlacion
corr_matrix = df3.corr().abs()

# Seleccione el triángulo superior de la matriz de correlación
superior = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(np.bool))

# indice de variables con correlación superir al 80%
v_eliminar = [column for column in superior.columns if any(superior[column] > 0.8)]
```

```python
v_eliminar
```

```python
# Drop features 
df4=DataFrame(df3.drop(df[v_eliminar], axis=1))
print("número de filas:", df4.shape[0], "número de columnas:", df4.shape[1])
```

```python
# valor medio y desviación estándard de todos los atributos numéricos:
df4.describe().loc[["mean", "std", "min","max"],:] 
```
